```python
import pandas as pd
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('sex')['survived'].sum()
```

```python
df.groupby('sex')['survived'].mean()
```

```python
df.groupby(['sex','embark_town'])['survived'].mean()
```

```python
df.groupby(['sex','embark_town'])['age'].mean()
```

```python
df.groupby(['sex','embark_town','who'])['survived'].mean()
```

```python
df.groupby(['sex','embark_town','who'])['survived'].mean()
```

```python
df.groupby(['sex','embark_town'])['age'].mean().unstack()
```

```python
df.pivot_table('survived',index='sex',columns=['embark_town'])
```

```python
df.pivot_table('survived',index='sex',columns=['embark_town'],aggfunc=['sum','mean','std'])
```

```python
df.pivot_table('survived',index=['sex','class'],columns=['embark_town'],aggfunc=['sum','mean','std'])
```
