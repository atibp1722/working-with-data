# Concat Dataframes

```python
df=pd.DataFrame({'A':['A0','A1','A2','A3'],
                 'B':['B0','B1','B2','B3'],
                 'C':['C0','C1','C2','C3'],
                 'D':['D0','D1','D2','D3']}
                )

df1=pd.DataFrame({'A':['A4','A5','A6','A7'],
                 'B':['B4','B5','B6','B7'],
                 'C':['C4','C5','C6','C7'],
                 'D':['D4','D5','D6','D7']}
                )

df2=pd.DataFrame({'A':['A8','A9','A10','A11'],
                 'B':['B8','B9','B10','B11'],
                 'C':['C8','C9','C10','C11'],
                 'D':['D8','D9','D10','D11']}
                )
df
df1
df2
```

```python
merged_df=pd.concat([df,df1,df2])
merged_df
```

```python
merged_df=pd.concat([df,df1,df2],ignore_index=True)
merged_df
```

```python
df1.columns=['A','B','C','E']
```

```python
pd.concat([df,df1],ignore_index=True)
```

```python
pd.concat([df,df1],ignore_index=True,join='inner')
```

```python
pd.concat([df,df1,df2],axis=1)
```

# Merge Dataframes

```python
df3=pd.DataFrame({'Name':['Anisha','Bill','Kumar','Gita','Lucy'],
                 'Department':['HR','IT','IT','ACC','MKT'],
                 'Designation':['MGR','OFF','DIR','OFF','OFF']}
                )

df4=pd.DataFrame({'Name':['Kumar','Gita','Lucy','Bill','Anisha'],
                 'Country':['USA','NEP','USA','CAN','NEP'],
                 'Year':['2020','2019','2020','2022','2021']}
                )
df3
df4
```

```python
pd.concat([df3,df4],axis=1)
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,on='Name',how='left')
```

```python
pd.merge(df3,df4,on='Name',how='right')
```

```python
df5=pd.DataFrame({'Director':['Sam','Reeta','Ayush','Nikita'],
                 'Department':['HR','IT','MKT','ACC']})
df5
```

```python
pd.merge(merge_df,df5,on='Department',how='right')
```

```python
data=pd.merge(merge_df,df5,on='Department',how='right')
```

# Join Dataframes

```python
dfa=pd.DataFrame({'A':['A0','A1','A2'],
                 'B':['B0','B1','B2']},
                 index=['A0','A1','A2'])

dfb=pd.DataFrame({'C':['C0','C2','C3'],
                 'D':['D0','D2','D3']},
                 index=['A0','A2','A3'])

dfa
dfb
```

```python
dfa.join(dfb)
```

```python
dfa.join(dfb,how='left')
```

```python
dfa.join(dfb,how='right')
```

```python
dfa.join(dfb,how='inner')
```

```python
dfa.join(dfb,how='outer')
```

```python
dfa.join(dfb,lsuffix='left',rsuffix='right',how='inner')
```
